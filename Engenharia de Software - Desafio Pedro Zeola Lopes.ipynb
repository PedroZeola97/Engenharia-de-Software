###Google Sheets Authentication

```
from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials

gc = gspread.authorize(GoogleCredentials.get_application_default())
```

###Adding Google Sheet

```
spreadsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1xAR2rSz9b8DHcvBLPItUAimEs9dVF7qFBakyqg_UtaY/edit#gid=0')
```

```
page = spreadsheets.sheet1
```

###Turning Into a DataFrame

```
import pandas as pd
dataframe = pd.DataFrame(page.get_all_values())
```

###Checking conditions

```

for i in range(3,len(dataframe)):
  faltas = (int(dataframe[2][i]))
  media = (int(dataframe[3][i]) + int(dataframe[4][i]) + int(dataframe[5][i]))/30
  if faltas > 15:
    dataframe[6][i] = 'Reprovado por Falta'
    dataframe[7][i] = 0
  elif media < 5:
    dataframe[6][i] = 'Reprovado por Nota'
    dataframe[7][i] = 0
  elif 5 <= media < 7:
    dataframe[6][i] = 'Exame Final'
    naf= (10 - media)
    dataframe[7][i] = naf
  else:
    dataframe[6][i] = 'Aprovado'
    dataframe[7][i] = 0
```
